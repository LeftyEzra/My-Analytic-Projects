```python
import pandas as pd
import requests 
from bs4 import BeautifulSoup
```

# Denver Nuggetts

```python
url = "https://www.statmuse.com/nba/ask/denver-nuggets-last-ten-games"
page = requests.get(url)
```

```python
denver_soup = BeautifulSoup(page.text, 'html')
```

```python
print(denver_soup.prettify())
```

denver_soup.find('table')

```python
denver_soup.find_all('table')[0]
```

```python
denver_table = denver_soup.find_all('table')[0]
```

```python
denver = denver_table.find_all('th')
denver
```

```python
game_column_titles = [title.text.strip() for title in denver]
game_column_titles
```

```python
import pandas as pd

denver_df = pd.DataFrame(columns= game_column_titles)
denver_df
```

```python
column_data = denver_table.find_all('tr')
for row in column_data[1:]:
    row_data = row.find_all("td") 
    row_values = [row_list.text.strip() for row_list in row_data]   
    print(row_values)   
     
        
     
    length= len(denver_df)   
     
    denver_df.loc[length] = row_values
```

```python
denver_df
```

```python
denver_df.to_csv("denver_nugget10.csv")
```

```python
denver_df = pd.read_csv('denver_nugget10.csv')
denver_df
```

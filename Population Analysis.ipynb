### Creating a treemap to view all the nations and their population

```python
treemap = df1[['Country', 'Capital', 'Continent',
                    'Population (2023)',  'Yearly Change',  'World Share' ]]
treemap
```

```python
treemap.loc[:, 'Year'] = 2023
treemap
```

```python
#Updating the missing continent of the countries in the DataFrame.
#Arguments
#treemap: The DataFrame containing rent data.
#country: The country name.
#continent: The name of the continent.

def update_continent(dataframe, country, continent):
   
    #querying the rows affected
    country_rows = treemap[treemap["Country"] == country]
    dataframe.loc[country_rows.index, 'Continent'] = continent

# Calling the function
update_continent(treemap, country="Caribbean Netherlands", continent='Europe')
update_continent(treemap, country="Congo", continent="Africa")
update_continent(treemap, country="Curaçao", continent='North America')
update_continent(treemap, country="Cabo Verde", continent='Africa')
update_continent(treemap, country="Czech Republic (Czechia)", continent='Europe')
update_continent(treemap, country="Côte d'Ivoire", continent='Africa')
update_continent(treemap, country="Faeroe Islands", continent='Europe')
update_continent(treemap, country="Holy See", continent='Europe')
update_continent(treemap, country="Macao", continent='Asia')
update_continent(treemap, country="Réunion", continent='Africa')
update_continent(treemap, country="Saint Helena", continent='Africa')
update_continent(treemap, country="Saint Kitts & Nevis", continent='North America')
update_continent(treemap, country="Saint Pierre & Miquelon", continent='North America')
update_continent(treemap, country="Sao Tome & Principe", continent='Africa')
update_continent(treemap, country="St. Vincent & Grenadines", continent='North America')
update_continent(treemap, country="State of Palestine", continent='Asia')
update_continent(treemap, country="Turks and Caicos", continent='North America')
update_continent(treemap, country="U.S. Virgin Islands", continent='North America')
update_continent(treemap, country="Wallis & Futuna", continent='Oceania')
```

```python
# Create the treemap figure
fig = px.treemap(
    treemap,
    path=[px.Constant("world"), 'Continent', 'Country'],
    values='Population (2023)',
    color="Population (2023)",
    hover_data=["World Share"],
    color_continuous_scale='RdBu',
    template="plotly_dark"
)

# Add direct labels to each rectangle
fig.update_traces(textinfo="label+value")



# Add a legend for the color scale
fig.update_layout(coloraxis_colorbar=dict(title="Population (2023)"))

# Set layout margins
fig.update_layout(margin=dict(t=50, l=25, r=25, b=25))

# Show the treemap
fig.show()
```

# Life Expectancy

This is the  average number of years that a new born could expect to live, if he or she were to pass through life exposed to sex- and age specific death rates prevailing at the time of his or her birth, for a specific year, in a given country, territory, or geographic area.

Or the average number of years a person is expected to live from birth.
                    
                    

```python
location_df = pd.read_csv("locations.csv")
```

```python
life_exp_df = location_df[['continent','location','life_expectancy']]
life_exp_df
```

```python
life_exp_df.info()
```

```python
stat_view = life_exp_df['life_expectancy'].describe().reset_index()
stat_view.columns = ['statistics', 'value']
stat_view
```

```python
fig = px.bar(life_exp_filtered, x='statistics', y='value',color='value',
             text='value', opacity=1)

#Customize the y-axis tick labels
fig.update_layout(
                    height=400,template='plotly_dark',
                    font=dict(family="Arial", size=13, color="white"),
                    xaxis_title="Life Expectancy",  # Added x-axis title
                    yaxis_title="Values",
                    title="Statitical view of life expectancy in the world".title(),
                    title_font=dict(size=18),
                    title_x=0.5, title_y=0.95, title_xanchor='center',
                    )
fig.update_traces(texttemplate=' %{text:.0f}', textfont_size=20, textsrc="white",textposition="outside",marker_color=['orange','red','purple','green'])
fig.update_yaxes(showgrid=False, range=[0, 100])

fig.show()
```

     - **Minimum (min)**: The lowest life expectancy is **53.
     - **50% (Median)**: The middle value when all life expectancies are sorted.
     - **Maximum (max)**: The highest life expectancy which  a person is expected to live is **87**.

# Low Life Expectancy: Causes, Effects & Solutions


A low life expectancy can be defined as a below-average age a person might reach on average when he or she is born.
In poor developing countries, life expectancy is quite low compared to rich industrialized countries.

According to WHO, the average life expectancy of the world is 73. It was also observed that Women have a higher life expectancy than Men. Women may live up to 74 years while men live around 70 years.

In this dataset, the region with the lowest average life expectancy is the Africa continent with a life expectancy as low as 53 years. While the region with the highest life expectancy is the European region with a life expectancy up to 86 years.

## Poverty:

**Poverty** is one major cause of a low life expectancy.

The less priviledge people have many demerits when it comes to medical treatment, education and several other factors that would contribute to a higher life expectancy.

(Moreso, the children of poor parents are more likely to have a low life expectancy since poverty can be inherited.) **God Forbid!!!**

### Homelessness:
If a person is  homeless, his or her life expectancy decreases as well.

Additionally, you have also a greater risk to become a victim to violence if you live on the streets since you often are unprotected against attacks.


### Fast food:
The intake of excessive amounts of fast food may also lower life expectancy significantly.

Fast food is often quite unhealthy and increases the risk for diabetes or heart conditions.

It also contributes to becoming obese and the resulting adverse health effects. Therefore, unhealthy food consumption can substantially increase the probability of serious health conditions and might therefore lower life expectancy.

### Alcohol:
Alcohol is another big threat when it comes to life expectancya and is often glamorized in our society.
Alcohol is one of the most dangerous drugs on earth.
Alcohol can lead to cirrhosis of the liver and may also lead to several other physical health problems.
Alcohol can also increase the risk for mental illness.

For people consuming excessive amounts of alcohol on a regular basis, they people have a much lower life expectancy compared to people who refrain from drinking too much alcohol.

### Tobacco:
Ministry of health warns that smokers are liable to die young.

Smoking tobacco is another significant contributor to a low life expectancy.

There are several health issues related to tobacco and smoking generally.

Some of them, like lung cancer, are quite deadly and significantly lower your life expectancy.

The excessive intake of tobacco or any other smoking means may also lead to additional health issues, like heart conditions or other serious conditions.

### Illegal drugs:
The use of illegal drugs may also lower your life expectancy and the fastest way to be mentally unstable.

Depending on the drug, it can cause severe physical health conditions.

Some drugs may also lead to an overdose, which in turn can translate into death for the respective person.

Moreover, the use of hard drugs also increases the likelihood to end up homeless, which further decreases life expectancy.

### Genetics:
Genetics can also play a major role for individual life expectancy.

If you have parents who lived quite long, chances are that your genetics will allow you to live longer as well.

Although the connection between genetics and life expectancy is quite hard to prove since there are so many other factors that come into play, many researchers have concluded that genetics play an important role for individual life expectancy.

### Stress:
Another contributor to low life expectancy is stress.

In our today society, people are often overwhelmed by the level of information or news that is reaching them.

Moreso, in most jobs, studies found that people feel stressed plenty of times.

Stress can drastically lower life expectancy since it leads to all kinds of health-related issues.For instance, stress may lead to hearth conditions and strokes. It may also lead to mental issues, which in turn may lead to an increased probability for accidents or suicide.

### Lack of exercising:
Physical activity is important to keep our bodies healthy.This does not necessarily mean that you have to go to the gym.
It could just mean to go for jogging or a walk every day.

Studies have shown that only 30 minutes of physical activity each day significantly improves your health.

Thus, in turn, a lack of exercise may lower your life expectancy since you are more likely to develop obesity or other diseases which in turn may lead to heart conditions and may eventually kill you.

### Lack of health insurance:
Life expectancy can also depend on the presence or absence of proper health insurance.

If you have no health insurance, the time may come when you suffer from a severe disease. Chances are that you will not be able to afford the treatment costs without health insurance.

Thus, if you do not get proper medical treatment, your chances to die from those kinds of serious diseases increase dramatically.

Thus, a lack of health insurance can contribute to a low life expectancy.

### Low quality of public infrastructure:
Depending on the country you are living in, there may be enormous differences in the quality of public infrastructure.

This includes the quality of public transportation, the quality of medical treatment and the likes and how accessible they are.

If you live in a country with low medical standards, your life expectancy might also be low since in case of serious health conditions, you might not get proper treatment.

### Diseases:
If one live in regions where many serious diseases are prevalent, your life expectancy will also be lower since you have a higher chance to die from those health issues compared to people living in countries where there are less serious diseases around.

### Parenting:
Parents have a big influence on their children.

If you grew up in a household where your parents treated you with respect and showed you how to navigate through your life in a safe manner, your life expectancy will be higher than if you have parents who have drug issues and do not take care of you properly.

Therefore, the kind and quality of parenting play a major role when it comes to accessing the life expectancy of people.

### Do more exercising:
Exercising is crucial in order to stay healthy and to increase your life expectancy.

This does not necessarily mean that you have to go to the gym every day and work out quite hard to kill yourself.

### Fight poverty at all cost:
Since the relationship betwixt poverty and low life expectancy are highly correllated, it is crucial to fight poverty.

This could be done using several different measures, including better education, higher minimum wages or additional measures to protect people and especially children to suffer from poverty and may be leaving your comfort zone.


### Integrate people in the society:
In order to increase life expectancy, it is also crucial that we integrate people in our society as well as possible and that no one is left behind.

### Change food consumption behavior:

```python
# Defining a function to webscrape data 
def data_scraping(url, index, date, opp, vs, win_lose):
    get_url = requests.get(url) # Make requests to get data.
    get_soup = BeautifulSoup(get_url.text, 'html') # Using beautiful soup to get and print the html file texts.
    
    # Confirmation to see if request is granted.
    data_table = get_soup.find_all('table')[index] # Filtering the tables. 
    data_table_header = data_table.find('thead') # Extract data for table header.
    
    data_table_header_values = data_table_header.find_all('tr') # Finding all table row in the theads
    for row in data_table_header_values:
        row_data = row.find_all("th")   
        data_table_header_row_values = [row_list.text.strip() for row_list in row_data] 
        #print(data_table_header_row_values) # This print statyement is for debugging purpose
        
        
    data_df = pd.DataFrame(columns= data_table_header_row_values) # Turn the extracted columns to a Pandas Dataframe.   
    
    
    # Getting the row values from the table.
    data_table_header_rows = data_table.find_all('tr')
    for row in data_table_header_rows[1:]:
        row_data = row.find_all("td")   
        row_values = [row_list.text.strip() for row_list in row_data]   
        #print(row_values) # For debugging purpose

        length= len(data_df)  

        data_df.loc[length] = row_values   
        
        
    # Appending some other columns into the dataframe.
    data_df['Date']  = date
    data_df['Opp']   = opp
    data_df['Hm/Aw'] = vs
    data_df['W/L']   = win_lose   
        
    return data_df   
scraped_data = data_scraping(den_LAC, 0, "2025-01-09", "LAC", "vs", "W" )   
       
```

```python
# Function to filter only row with Denver Nuggets
def denver_nuggets_stats(index, points):
    
    den_df = scraped_data[scraped_data['PLAYER'].isin(["Denver Nuggets"])]
    
    den_df = pd.DataFrame(den_df)

    den_df.loc[index, "Opp_PTS"] = points 
   
    
    den_df['2M-2A'] = den_df['2M-2A'].str.replace('-', ' ')
    den_df['3M-3A'] = den_df['3M-3A'].str.replace('-', ' ')
    den_df['1M-1A'] = den_df['1M-1A'].str.replace('-', ' ')
    #den_df['FG%']   = den_df['FG%'].str.replace('%', '')
    
    
    den_df[["2PM","2PA"]] = den_df["2M-2A"].str.rsplit(" ", n=1, expand=True)
    den_df[["3PM","3PA"]] = den_df["3M-3A"].str.rsplit(" ", n=1, expand=True)
    den_df[["1M","1A"]] = den_df["1M-1A"].str.rsplit(" ", n=1, expand=True)
    
    den_df.loc[:, 'Date'] = pd.to_datetime(den_df['Date'])
    den_df['Date'] = den_df['Date'].map(lambda x: x.strftime('%Y-%m-%d'))
    
    den_df[[ "2PA", '2PM', "3PA", "3PM", "1M", "1A",  "Or", "Dr", "To", "Stl", "Blk", "Fo", "+/-", "Eff"]] = den_df[["2PA", '2PM', "3PA", "3PM", "1M", "1A", "Or", "Dr", "To", "Stl", "Blk", "Fo", "+/-", "Eff"]].astype(int)
    
    den_df["FGM"] =   den_df["2PM"] + den_df["3PM"] 
    den_df["FGA"] =   den_df["2PA"] + den_df["3PA"] 
    
    den_df = den_df.loc[:,~den_df.columns.duplicated()]
    den_df = pd.DataFrame(den_df)
    
    den_df[["Pts", "Reb","Ast"]] = den_df[["Pts", "Reb","Ast"]].apply(pd.to_numeric)
    
    den_df = den_df.rename(columns={"1M" : "FTM","1A":"FTA","Reb":"REB","Ast":"AST","Or":"OREB","Dr":"DREB","To":"TOV","Stl":"STL","Blk":"BLK","Fo":"PF","Eff":"EFF"})
    
    denver2024_2025_basketball_stats = den_df[["Date","Hm/Aw","Opp","W/L","Pts","Opp_PTS", "PLAYER", "FGM","FGA","FG%","3PM","3PA","2PM","2PA","FTM","FTA","OREB","DREB","REB","AST","STL","BLK","TOV","PF","+/-","EFF"]]
    
    read_csv = pd.read_csv('2024-2025 Denver Nuggets Basketball Stats.csv')
    copy_csv = read_csv.copy()
    
    concat_csvs = pd.concat([copy_csv, denver2024_2025_basketball_stats ]).reset_index(drop=True)
    
    concat_csvs.to_csv('2024-2025 Denver Nuggets Basketball Stats.csv', index=False)
    
    return concat_csvs
```

```python
import pandas as pd
import requests 
from bs4 import BeautifulSoup
```

# Denver Nuggetts Games from Stat Muse

```python
#Copied url
url = "https://www.statmuse.com/nba/ask/denver-nuggets-last-ten-games"
page = requests.get(url)
```

```python
# Create a BeautifulSoup object to parse the HTML content
denver_soup = BeautifulSoup(page.text, 'html')
```

```python
# Display the prettified HTML content of the denver_soup object
print(denver_soup.prettify())
```

denver_soup.find('table')

```python
# Retrieve the first table element from the parsed HTML
denver_soup.find_all('table')[0]
```

```python
# Passing a variable to the retrieved table
denver_table = denver_soup.find_all('table')[0]
```

denver = denver_table.find_all('th')
denver

```python
# This line extracts column titles from the 'denver' variable
game_column_titles = [title.text.strip() for title in denver]
game_column_titles
```

```python
import pandas as pd

denver_df = pd.DataFrame(columns= game_column_titles) # Turn the extracted columns to a Pandas Dataframe
denver_df
```

```python
column_data = denver_table.find_all('tr') # Find all rows in the 'denver_table'
for row in column_data[1:]:# Iterate through each row (excluding the header row)
    row_data = row.find_all("td")    # Find all data cells (td) within the row
    row_values = [row_list.text.strip() for row_list in row_data]   # Extract the text from each cell and strip any leading/trailing whitespace
    print(row_values)    # Print the extracted values for this row
     
        
     
    length= len(denver_df)   # Get the current length of the 'denver_df' DataFrame
     
    denver_df.loc[length] = row_values  # Add the row values to the DataFrame as a new row
```

```python
denver_df # Display output
```

```python
denver_df.to_csv("denver_nugget10.csv") # Convert to csv or any file of your choice
```

```python
denver_df = pd.read_csv('denver_nugget10.csv')# Use pandas to read the file
denver_df
```

**What are Pandas dataframes?**

1. One way to fix a wrong data is by replacing them with a value: this error might be caused from typo error
2. Replace it with the average of the entire column
3. Replace it with the average of the values on the previous & next date

## What are the benefits of using the datetime data type instead of object?

Using Pandas DateTimeType instead of Object for representing dates and times in pandas DataFrame offers several benefits. 

**Enhance functionality:** There are some builtin functions Pandas datetime possesses for performing operations on date and time such as(year, month, day) calculating the difference between dates and times.

**Improved sorting** DateTimeDataType allows for easier sorting and filtering of data.

## How do you extract different parts of a date column like the month, year, month, weekday, etc., into separate columns? Illustrate with an example.

**We can now extract different parts of the data into separate columns, using the DatetimeIndex class**

```python
covid_df['year']    = pd.DatetimeIndex(covid_df.date).year
covid_df['month']   = pd.DatetimeIndex(covid_df.date).month
covid_df['day']     = pd.DatetimeIndex(covid_df.date).day
covid_df['weekday'] = pd.DatetimeIndex(covid_df.date).weekday
covid_df['hour'] = pd.DatetimeIndex(covid_df.date).hour
```

**average():** same as mean

**describe():** return the statistics summary of each group

**max():** returns the maximum number for each group

**min():** returns the manimum number for each group

**first():** returns the first value for each group

**last():** returns the last value for each group

## What do you mean by a running or cumulative sum?

Running total or cumulative sums are used to display the total sum of data as it grows with time.

## How do you create a new column containing the running or cumulative sum of another column?

```python
# Creating a different column for the total number of cases
covid_df['total_cases'] = covid_df.new_cases.cumsum()
```

## What are other cumulative measures supported by Pandas dataframes?

**Some of the free cumlative measures supported by Pandas include:**

**.cumsum():** Calculates the cumulative sum of a column or series, returning a new seires with thr=e running total.

**.cumprod():** calculates the cumulative product of a column or series , returning a new series with the running product.

**.cummax() and cummin():** calculates the cumulative maximum and minimum values of a column or series, and returning a new series with the running maximum or minimum.

**.cumcount():**  calculates the cumulative count of non-null values in a column and returning a new series with the running count.

## What does it mean to merge two dataframes? Give an example.

**Merging 2 dataframes refers to combining or joining them into a single dataframe based on common column or index.**

The resulting merged dataframe contains the data from both original dataframes and are aligned based on specified columns or index
